```python
import pandas as pd
import numpy as np
import random
import matplotlib.pyplot as plt
from sklearn.preprocessing import OneHotEncoder
```

```python
insu = pd.read_csv('insurance.csv')  
W = np.zeros(6)
for i in range(len(W)):
    W[i] = random.randint(1,100)
cols = list(insu.columns)
print(insu.where((insu['region'] == 'southwest') & (insu['smoker'] == 'no')).dropna())
plt.scatter(insu['bmi'].where(insu['smoker'] == 'yes').dropna(),insu['charges'].where(insu['smoker'] == 'yes').dropna(),label='Smoker Insurance Charges',color='blue');
plt.scatter(insu['bmi'].where(insu['smoker'] == 'no').dropna(),insu['charges'].where(insu['smoker'] == 'no').dropna(),label='Non Smoker Insurance Charges',color='red');
plt.scatter(insu['bmi'].where(insu['sex'] == 'male').dropna(),insu['charges'].where(insu['sex'] == 'male').dropna(),label='Male Insurance Charges',color='black');
plt.scatter(insu['bmi'].where(insu['sex'] == 'female').dropna(),insu['charges'].where(insu['sex'] == 'female').dropna(),label='Female Insurance Charges',color='green');

```
